Creating  DataFrames

1. FROM RDD's

```

rdd=sc.parallelize([['2017-01-01','Rachel',19,156,'Sydney'],
     ['2018-01-01','Steve',20,160,'Melbourne'],
     ['2019-01-01','Robert',29,146,'Brisbane'],
     ['2020-01-01','Michael',31,176,'Sydney']
    ])
```

```
simple_df = spark.createDataFrame(rdd,["Date","Name","Age","No","Location"])
```

```
simple_df.show()
```

```
rdd.take(4)
```

From Json String

```
json_string=[
  '{"Date":"2017-02-01","Name":"Rachel","Age":19,"Weight":156,"Location":"Sydney"}',
  '{"Date":"2017-08-01","Name":"Ravi","Age":29,"Weight":87,"Location":"Newyork"}',
  '{"Date":"2017-04-01","Name":"Robert","Age":9,"Weight":72,"Location":"Los Angeles"}',
  '{"Date":"2017-02-01","Name":"Rahim","Age":11,"Weight":55,"Location":"Brisbane"}'
]

simple_df_json=spark.read.json(sc.parallelize(json_string))
simple_df_json.show()
```
