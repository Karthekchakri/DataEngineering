Convert list of dicts to sparkdataframe using Row

```
users_list=[
  {'user_id':1,'user_first_name':'Scott'},
  {'user_id':2,'user_first_name':'Donald'},
  {'user_id':3,'user_first_name':'Mickey'},
  {'user_id':4,'user_first_name':'Elvis'}
]
```

```
type(users_list)
```

```
user_rows=[Row(*users.values()) for users in users_list]
```

```
user_rows
```

```
df =spark.createDataFrame(user_rows,'user_id int , user_name string')
```

```
df
```

```
user_rows=[Row(**users) for users in users_list]
```

```
user_rows
```

```
df =spark.createDataFrame(user_rows,'user_id int , user_first_name string')
```

```
df.show()
```
